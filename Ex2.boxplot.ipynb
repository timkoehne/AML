# Load Datasets

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "burnoutminer/heights-and-weights-dataset",
   "SOCR-HeightWeight.csv",
)

dataset_0 = df["Height(Inches)"]
dataset_1 = df["Weight(Pounds)"]


df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
     "andrewmvd/heart-failure-clinical-data",
     "heart_failure_clinical_records_dataset.csv"
)

dataset_2 = df["creatinine_phosphokinase"]
```

# Task 4: Create Boxplots for the data from Exercise 1


```python
sns.boxplot(data=dataset_0, orient="h")
plt.show()
sns.boxplot(data=dataset_1, orient="h")
plt.show()
sns.boxplot(data=dataset_2, orient="h")
plt.show()
```

# Task 5: Interpret the Boxplots
The median is the center-line. The box shows the ICR (interquartile range) from Q1 at 25% (lower quartile) to Q3 at 75% (upper quartile) of the data.
The whiskers seperates normal from anormal data. They are calculated as Q1 - 1.5\*IQR and Q3 + 1.5\*IQR.
